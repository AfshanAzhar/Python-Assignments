### Assignment 1: Iris dataset Train Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
```

```python
import numpy as np
import pandas as pd
```

```python
iris = pd.read_csv("data/iris.csv")
```

```python
iris.head()
```

```python
iris.shape
```

```python
iris = iris[['sl','sw', 'pl', 'pw', 'class']] # removing serial column
```

```python
iris.head()
```

```python
iris.shape
```

### Training data 

```python
# Shuffling and diving the data 
train = iris.sample(frac= 0.7)
test = iris.sample(frac = 0.3)
```

```python
train.shape
```

```python
test.shape
```

```python
train['class'].value_counts()
```

```python
test['class'].value_counts()
```

```python
# now diving the dataset into inputs and outputs
```

```python
x_train = train[['sl', 'sw', 'pl', 'pw']]
y_train = train[['class']]
x_test = test[['sl', 'sw', 'pl', 'pw']]
y_test = test[['class']]
```

```python
#one hot vector encoding 
y_train = pd.get_dummies(y_train)
y_test = pd.get_dummies(y_test)
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
# building model layers for training
```

```python
model = Sequential()
model.add(Dense(10, activation = 'relu', input_shape = (4,)))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(3, activation='softmax'))
```

```python
#compiling....
```

```python
model.compile(optimizer = SGD(lr = 0.002),
             loss = "categorical_crossentropy",
             metrics = ['acc'])
```

```python
model.fit(x_train, y_train, epochs = 450)
```

```python
# here we got 99 % accuracy
```

### Testing

```python
model.evaluate(x_test, y_test)
```
